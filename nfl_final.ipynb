```python
#gives the week in which the team didn't play for that season or year

def byeweek(team, year):
    gameweeks = []
    for x in range(4, 13):
        games = nflgame.live._games_in_week(year=year,week=x,kind='REG')
        if games == []:
            return 0
        weekteams = []
        for i in games:
            weekteams.append(i["home"])
            weekteams.append(i["away"])
        if team not in weekteams and weekteams!=[]:
            return x
    return 0
```

```python
#gives the average stats summing the previous weeks

def stats_avg_home_test(y,w,h,k,d):
    t = 0
    g = 0
    bye = byeweek(team=h,year=y)
    rush_yds_s = 0
    rush_yds_avg_s = 0
    pass_yds_avg_s = 0
    pass_yds_s = 0
    pen_avg_s = 0
    pen_avgt_s = 0
    punt_avgt_s = 0
    punt_avg_s = 0
    interception_s = 0
    fumbles_s = 0
    sack_s = 0
    sack_avg_s = 0
    rush_yds_a = 0
    rush_yds_avg_a = 0
    pass_yds_avg_a = 0
    pass_yds_a = 0
    pen_avg_a = 0
    pen_avgt_a = 0
    punt_avgt_a = 0
    punt_avg_a = 0
    interception_a = 0
    fumbles_a = 0
    sack_a = 0
    sack_avg_a = 0
    turnover_diff = 0
        
        
    while(t<w):
        week_avg = nflgame.one(y,t,h,h,k)
        if week_avg is not None:
            game = nflgame.games(year=y,week=t,home=h,away=h,kind=k)
            players = nflgame.combine_game_stats(game)
            
            if h == week_avg.home:
                rush_yds_s += float(week_avg.stats_home.rushing_yds)
                rush_yds_a += float(week_avg.stats_away.rushing_yds)
                pass_yds_s += float(week_avg.stats_home.passing_yds)
                pass_yds_a += float(week_avg.stats_away.passing_yds)
                punt_avgt_s += float(week_avg.stats_home.punt_avg)
                punt_avgt_a += float(week_avg.stats_away.punt_avg)
                if week_avg.stats_home.penalty_cnt!= 0:
                    pen_avgt_s += float(week_avg.stats_home.penalty_yds/week_avg.stats_home.penalty_cnt)
                else:
                    pen_avgt_s += 0
                if week_avg.stats_away.penalty_cnt!=0:
                    pen_avgt_a += float(week_avg.stats_away.penalty_yds/week_avg.stats_away.penalty_cnt)
                else:
                    pen_avgt_a +=0
            else:
                rush_yds_s += float(week_avg.stats_away.rushing_yds)
                rush_yds_a += float(week_avg.stats_home.rushing_yds)
                pass_yds_s += float(week_avg.stats_away.passing_yds)
                pass_yds_a += float(week_avg.stats_home.passing_yds)
                punt_avgt_s += float(week_avg.stats_away.punt_avg)
                punt_avgt_a += float(week_avg.stats_home.punt_avg)
                if week_avg.stats_away.penalty_cnt!= 0:
                    pen_avgt_s += float(week_avg.stats_away.penalty_yds/week_avg.stats_away.penalty_cnt)
                else:
                    pen_avgt_s += 0
                if week_avg.stats_home.penalty_cnt!=0:
                    pen_avgt_a += float(week_avg.stats_home.penalty_yds/week_avg.stats_home.penalty_cnt)
                else:
                    pen_avgt_a +=0

            #turnover differential for the home team
            for i in players.defense():
                if i.team == h:
                    interception_s+=float(i.defense_int)
                else:
                    interception_a+=float(i.defense_int)
            for f in players.fumbles():
                if f.team == h:
                    fumbles_a+= float(f.fumbles_lost)
                else:
                    fumbles_s+= float(f.fumbles_lost)
                
            #sack scored and allowed by the home team
            for s in players.defense():
                if s.team == h:
                    sack_s += float(s.defense_sk)
                else:
                    sack_a += float(s.defense_sk)
        t= t+1
    if bye<w:
        g = 1
    rush_yds_avg_s = float(rush_yds_s/(w-g-1))
    pass_yds_avg_s = float(pass_yds_s/(w-g-1))
    pen_avg_s = float(pen_avgt_s/(w-g-1))
    punt_avg_s = float(punt_avgt_s/(w-g-1))
    sack_avg_s = float(sack_s/(w-g-1))
    rush_yds_avg_a = float(rush_yds_a/(w-g-1))    
    pass_yds_avg_a = float(pass_yds_a/(w-g-1))
    pen_avg_a = float(pen_avgt_a/(w-g-1))
    punt_avg_a = float(punt_avgt_a/(w-g-1))
    sack_avg_a = float(sack_a/(w-g-1))
    turnover_diff = float(((interception_s+fumbles_s)-(interception_a+fumbles_a))/(w-g-1))
        
    d = d.append({"hRushYdsScored":rush_yds_avg_s,"hPassYdsScored":pass_yds_avg_s,"hPuntAvgScored":punt_avg_s,
                  "hPenAvgScored":pen_avg_s,"hSackAvgScored":sack_avg_s,"hRushYdsAllowed":rush_yds_avg_a,
                  "hPassYdsAllowed":pass_yds_avg_a,"hPuntAvgAllowed":punt_avg_a,"hPenAvgAllowed":pen_avg_a,
                  "hSackAvgAllowed":sack_avg_a,"hTurnoverDiff":turnover_diff},ignore_index =True)
    return d
```

```python
def stats_avg_away_test(y,w,h,k,d):
    t = 0
    g = 0
    bye = byeweek(h,y)
    rush_yds_s = 0
    rush_yds_avg_s = 0
    pass_yds_avg_s = 0
    pass_yds_s = 0
    pen_avg_s = 0
    pen_avgt_s = 0
    punt_avgt_s = 0
    punt_avg_s = 0
    interception_s = 0
    fumbles_s = 0
    sack_s = 0
    sack_avg_s = 0
    rush_yds_a = 0
    rush_yds_avg_a = 0
    pass_yds_avg_a = 0
    pass_yds_a = 0
    pen_avg_a = 0
    pen_avgt_a = 0
    punt_avgt_a = 0
    punt_avg_a = 0
    interception_a = 0
    fumbles_a = 0
    sack_a = 0
    sack_avg_a = 0
    turnover_diff = 0    
        
    while(t<w):
        week_avg = nflgame.one(y,t,h,h,k)
        if week_avg is not None:
            game = nflgame.games(year=y,week=t,home=h,away=h,kind=k)
            players = nflgame.combine_game_stats(game)
            
            if h== week_avg.home:  
                rush_yds_s += float(week_avg.stats_home.rushing_yds)
                rush_yds_a += float(week_avg.stats_away.rushing_yds)
                pass_yds_s += float(week_avg.stats_home.passing_yds)
                pass_yds_a += float(week_avg.stats_away.passing_yds)
                punt_avgt_s += float(week_avg.stats_home.punt_avg)
                punt_avgt_a += float(week_avg.stats_away.punt_avg)
                if week_avg.stats_home.penalty_cnt!= 0:
                    pen_avgt_s += float(week_avg.stats_home.penalty_yds/week_avg.stats_home.penalty_cnt)
                else:
                    pen_avgt_s += 0
                if week_avg.stats_away.penalty_cnt!=0:
                    pen_avgt_a += float(week_avg.stats_away.penalty_yds/week_avg.stats_away.penalty_cnt)
                else:
                    pen_avgt_a +=0
            else:
                rush_yds_s += float(week_avg.stats_away.rushing_yds)
                rush_yds_a += float(week_avg.stats_home.rushing_yds)
                pass_yds_s += float(week_avg.stats_away.passing_yds)
                pass_yds_a += float(week_avg.stats_home.passing_yds)
                punt_avgt_s += float(week_avg.stats_away.punt_avg)
                punt_avgt_a += float(week_avg.stats_home.punt_avg)
                if week_avg.stats_away.penalty_cnt!= 0:
                    pen_avgt_s += float(week_avg.stats_away.penalty_yds/week_avg.stats_away.penalty_cnt)
                else:
                    pen_avgt_s += 0
                if week_avg.stats_home.penalty_cnt!=0:
                    pen_avgt_a += float(week_avg.stats_home.penalty_yds/week_avg.stats_home.penalty_cnt)
                else:
                    pen_avgt_a +=0
                
            #turnover differential of the away team
            for i in players.defense():
                if i.team == h:
                    interception_s+=float(i.defense_int)
                else:
                    interception_a+=float(i.defense_int)
            for f in players.fumbles():
                if f.team == h:
                    fumbles_a+= float(f.fumbles_lost)
                else:
                    fumbles_s+= float(f.fumbles_lost)
                    
            #sacks scored and allowed by the away team
            for s in players.defense():
                if s.team == h:
                    sack_s += float(s.defense_sk)
                else:
                    sack_a += float(s.defense_sk)
        t= t+1
    if bye<w:
        g = 1
    rush_yds_avg_s = float(rush_yds_s/(w-g-1))    
    pass_yds_avg_s = float(pass_yds_s/(w-g-1))    
    pen_avg_s = float(pen_avgt_s/(w-g-1))
    punt_avg_s = float(punt_avgt_s/(w-g-1))    
    sack_avg_s = float(sack_s/(w-g-1))
    rush_yds_avg_a = float(rush_yds_a/(w-g-1))
    pass_yds_avg_a = float(pass_yds_a/(w-g-1))    
    pen_avg_a = float(pen_avgt_a/(w-g-1))
    punt_avg_a = float(punt_avgt_a/(w-g-1))
    sack_avg_a = float(sack_a/(w-g-1))    
    turnover_diff = float(((interception_s+fumbles_s)-(interception_a+fumbles_a))/(w-g-1))
        
    d = d.append({"aRushYdsScored":rush_yds_avg_s,"aPassYdsScored":pass_yds_avg_s,"aPuntAvgScored":punt_avg_s,
                  "aPenAvgScored":pen_avg_s,"aSackAvgScored":sack_avg_s,"aRushYdsAllowed":rush_yds_avg_a,
                  "aPassYdsAllowed":pass_yds_avg_a,"aPuntAvgAllowed":punt_avg_a,"aPenAvgAllowed":pen_avg_a,
                  "aSackAvgAllowed":sack_avg_a,"aTurnoverDiff":turnover_diff},ignore_index =True)
    return d
```

```python
def week_stats_avg_test(y,w,k,d):
    df_home = pd.DataFrame(columns = ["hRushYdsScored","hPassYdsScored","hPuntAvgScored",
                                   "hPenAvgScored","hSackAvgScored","hRushYdsAllowed","hPassYdsAllowed",
                                   "hPuntAvgAllowed","hPenAvgAllowed","hSackAvgAllowed",'hTurnoverDiff'])
    df_away = pd.DataFrame(columns = ["aRushYdsScored","aPassYdsScored","aPuntAvgScored","aPenAvgScored",
                                   "aSackAvgScored","aRushYdsAllowed","aPassYdsAllowed","aPuntAvgAllowed",
                                   "aPenAvgAllowed","aSackAvgAllowed",'aTurnoverDiff'])
    games = nflgame.live._games_in_week(year=y,week=w,kind=k)
    for x in games:
        df_home = stats_avg_home_test(y,w,x['home'],k,df_home)
        df_away = stats_avg_away_test(y,w,x['away'],k,df_away)
        d = d.append({'Year':y,"Week":w,"Home":x['home'],"Away":x['away']},ignore_index = True)
    d = d.dropna(axis=1)
    d = pd.concat([d,df_home,df_away],axis=1)
    return d
```

```python
def week_stats_past(y,w,k,d):
    df_home = pd.DataFrame(columns = ["hRushYdsScored","hPassYdsScored","hPuntAvgScored",
                                   "hPenAvgScored","hSackAvgScored","hRushYdsAllowed","hPassYdsAllowed",
                                   "hPuntAvgAllowed","hPenAvgAllowed","hSackAvgAllowed",'hTurnoverDiff'])
    df_away = pd.DataFrame(columns = ["aRushYdsScored","aPassYdsScored","aPuntAvgScored","aPenAvgScored",
                                   "aSackAvgScored","aRushYdsAllowed","aPassYdsAllowed","aPuntAvgAllowed",
                                   "aPenAvgAllowed","aSackAvgAllowed",'aTurnoverDiff'])
    games = nflgame.games(y,w,kind=k)
    for x in games:
        win=0
        if x.home == x.winner:
            win=1
        df_home = stats_avg_home_test(y,w,x.home,k,df_home)
        df_away = stats_avg_away_test(y,w,x.away,k,df_away)
        d = d.append({'Year':y,"Week":w,"Home":x.home,"Away":x.away,
                      'hScore':x.score_home,'aScore':x.score_away,'Win':win},ignore_index = True)
    d = d.dropna(axis=1)
    d = pd.concat([d,df_home,df_away],axis=1)
    return d
```
